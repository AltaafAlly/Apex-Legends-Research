```python
import pandas as pd

# Define legend categories
legend_categories = {
    'Assault': ['Bangalore', 'Fuse', 'Ash', 'Mad Maggie', 'Ballistic'],
    'Skirmisher': ['Pathfinder', 'Wraith', 'Octane', 'Revenant', 'Horizon', 'Valkyrie', 'Alter'],
    'Recon': ['Bloodhound', 'Crypto', 'Seer', 'Vantage'],
    'Support': ['Gibraltar', 'Lifeline', 'Mirage', 'Loba', 'Newcastle', 'Conduit'],
    'Controller': ['Caustic', 'Wattson', 'Rampart', 'Catalyst']
}

# Function to categorize legends
def categorize_legend(legend):
    for category, legends in legend_categories.items():
        if legend in legends:
            return category
    return 'Unknown'  # In case a legend does not fit into any category
```

```python
# Load datasets for PC, PS4, and Xbox
pc_legend_stats_df = pd.read_csv(r"C:\Users\altaa\Documents\GitHub\Apex-Legends-Research\Data_Retrieval\CSV_files\Player Stats\Cleaned Player Stats\Legend Stats\Legend_Stats_PC.csv")
ps4_legend_stats_df = pd.read_csv(r"C:\Users\altaa\Documents\GitHub\Apex-Legends-Research\Data_Retrieval\CSV_files\Player Stats\Cleaned Player Stats\Legend Stats\Legend_Stats_PS4.csv")
xbox_legend_stats_df = pd.read_csv(r"C:\Users\altaa\Documents\GitHub\Apex-Legends-Research\Data_Retrieval\CSV_files\Player Stats\Cleaned Player Stats\Legend Stats\Legend_Stats_Xbox.csv")

# Add 'legend_category' column based on the legend's name
pc_legend_stats_df['legend_category'] = pc_legend_stats_df['legend'].apply(categorize_legend)
ps4_legend_stats_df['legend_category'] = ps4_legend_stats_df['legend'].apply(categorize_legend)
xbox_legend_stats_df['legend_category'] = xbox_legend_stats_df['legend'].apply(categorize_legend)

# Display the first few rows of each DataFrame to verify the changes
pc_legend_stats_df.head(15)
```

```python
# Function to filter out the best legends based on stats
def filter_best_legends(df):
    # Ensure columns are numeric
    df['BR Kills'] = pd.to_numeric(df['BR Kills'], errors='coerce')
    df['BR Damage'] = pd.to_numeric(df['BR Damage'], errors='coerce')
    df['BR Wins'] = pd.to_numeric(df['BR Wins'], errors='coerce')
    
    # Remove legends with all stats (BR Kills, BR Damage, BR Wins) set to zero
    filtered_df = df[(df['BR Kills'] != 0) | (df['BR Damage'] != 0) | (df['BR Wins'] != 0)]
    
    # For each player, keep only the legends with some stats
    grouped = filtered_df.groupby('player_name', group_keys=False)
    
    # Sort values by multiple columns within each group
    best_legends = grouped.apply(lambda x: x.sort_values(['BR Kills', 'BR Damage', 'BR Wins'], ascending=[False, False, False]))
    
    return best_legends

# Apply filtering to each dataset
filtered_pc_legend_stats_df = filter_best_legends(pc_legend_stats_df)
filtered_ps4_legend_stats_df = filter_best_legends(ps4_legend_stats_df)
filtered_xbox_legend_stats_df = filter_best_legends(xbox_legend_stats_df)

# Display first few rows of filtered dataframes to verify
print("Filtered PC Legend Stats:")
filtered_pc_legend_stats_df.head(15)

# print("\nFiltered PS4 Legend Stats:")
# filtered_ps4_legend_stats_df.head()

# print("\nFiltered Xbox Legend Stats:")
# print(filtered_xbox_legend_stats_df.head())
```
