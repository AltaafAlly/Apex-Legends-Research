# General understanding of your data distribution for kills, revives, and wins across all players and platforms.

```python
import pandas as pd
# Replace 'your_dataset.csv' with the path to your actual dataset file


# Read CSV files for each platform
pc_data = pd.read_csv(r"C:\Users\altaa\Documents\GitHub\Apex-Legends-Research\Data_Retrieval\py_files\Notebook_For_Data_Preprocessing\Career Stats\Data Imputation Files PC\Career_Stats_PC_Final_Dataset.csv")
playstation_data = pd.read_csv(r'C:\Users\altaa\Documents\GitHub\Apex-Legends-Research\Data_Retrieval\py_files\Notebook_For_Data_Preprocessing\Data Imputation FIles PS and Xbox\PS\Career_Stats_mice_imputed_PS.csv')
xbox_data = pd.read_csv(r'C:\Users\altaa\Documents\GitHub\Apex-Legends-Research\Data_Retrieval\py_files\Notebook_For_Data_Preprocessing\Data Imputation FIles PS and Xbox\Xbox\Career_Stats_mice_imputed_Xbox.csv')
```

```python
# PC Platform Statistics
print("Descriptive Statistics for PC:")
pc_stats = pc_data[['career_kills', 'career_revives', 'career_wins']].describe()
display(pc_stats)

# PlayStation Platform Statistics
print("Descriptive Statistics for PlayStation:")
playstation_stats = playstation_data[['career_kills', 'career_revives', 'career_wins']].describe()
display(playstation_stats)

# Xbox Platform Statistics
print("Descriptive Statistics for Xbox:")
xbox_stats = xbox_data[['career_kills', 'career_revives', 'career_wins']].describe()
display(xbox_stats)

```

```python
# Dictionary to store mean values for each platform
platform_means_dict = {
    'PC': pc_data[['career_kills', 'career_revives', 'career_wins']].mean(),
    'PlayStation': playstation_data[['career_kills', 'career_revives', 'career_wins']].mean(),
    'Xbox': xbox_data[['career_kills', 'career_revives', 'career_wins']].mean()
}

# Convert the dictionary into a DataFrame
platform_means = pd.DataFrame(platform_means_dict).T.reset_index()
platform_means.rename(columns={'index': 'platform'}, inplace=True)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set up the matplotlib figure with subplots
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

# Bar plot for Career Kills by Platform
sns.barplot(x='platform', y='career_kills', data=platform_means, ax=axes[0], palette='Blues_d')
axes[0].set_title('Average Career Kills by Platform')
axes[0].set_xlabel('Platform')
axes[0].set_ylabel('Average Career Kills')

# Bar plot for Career Revives by Platform
sns.barplot(x='platform', y='career_revives', data=platform_means, ax=axes[1], palette='Greens_d')
axes[1].set_title('Average Career Revives by Platform')
axes[1].set_xlabel('Platform')
axes[1].set_ylabel('Average Career Revives')

# Bar plot for Career Wins by Platform
sns.barplot(x='platform', y='career_wins', data=platform_means, ax=axes[2], palette='Reds_d')
axes[2].set_title('Average Career Wins by Platform')
axes[2].set_xlabel('Platform')
axes[2].set_ylabel('Average Career Wins')

# Adjust layout
plt.tight_layout()

# Display the plots
plt.show()
```
